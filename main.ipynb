Implacing __mean__ of `rating` by group of `is_sponsored & sustainability_badges`

```python
rating_cleaned=df.groupby(['is_sponsored','sustainability_badges'])['rating'].transform(lambda x: round(x.mean(),1))
df['rating']=df['rating'].fillna(rating_cleaned)
```

Cleaning `number_of_reviews`

```python
df['number_of_reviews']=df['number_of_reviews'].str.replace(',','',regex=True)
```

```python
df['number_of_reviews']=df['number_of_reviews'].astype(float)
```

Finding best method to impute __nulls__ form `number_of_reviews`

```python
print(df['number_of_reviews'].value_counts())
print(f"Mean: {df['number_of_reviews'].mean()}")
print(f"Median: {df['number_of_reviews'].median()}")
```

```python
df.groupby(['rating','is_sponsored'])['number_of_reviews'].transform(lambda x: np.round(x.median(),1))
```

__mode__ is skewing values instead using groupby __median__ of `rating` & `is_sponsored` to fill __nulls__

```python
review_cleaning=df.groupby(['rating','is_sponsored'])['number_of_reviews'].transform(lambda x: np.round(x.median(),1))
df['number_of_reviews']=df['number_of_reviews'].fillna(review_cleaning)
```

```python
df['number_of_reviews']=df['number_of_reviews'].astype(int)
```

column `bought_in_last_month` is special because the number of buyers are corelated with reviews  
* first step is to clean `bought_in_last_month` to float
* applying groupby `number_of_reviews` to implace nulls by __median__ of `bought_in_last_month`

```python
df['bought_in_last_month']=df['bought_in_last_month'].str.extract(r'(^\d+K?)')
```

```python
def fixer (value):
    if 'K' in str(value):
        return float(value.replace('K',''))*1000
    else:
        return float(value)
```

```python
df['bought_in_last_month']=df['bought_in_last_month'].apply(fixer)
```

```python
df['bought_in_last_month']=df['bought_in_last_month'].astype('float')
```

applying `.median()` groupby  
> why `.median()` because `.mean()` would skew accuracy due to big range of reviews

```python
filled=df.groupby('number_of_reviews')['bought_in_last_month'].transform(lambda x: x.median())
```

```python
df['bought_in_last_month']=df['bought_in_last_month'].fillna(filled)
```

```python
nullg=df[df['bought_in_last_month'].isna()].groupby('number_of_reviews').size()
print(nullg)
```

aggregating all groups to find which groups have 0 count

```python
gs=df.groupby('number_of_reviews')['bought_in_last_month'].agg(['count','size'])
ang=gs[gs['count']==0]
print(ang)
```

Still some __nulls__ where found because some groups where empty so im implacing __nulls__ by __median__ of ``bought_in_last_month``

```python
df['bought_in_last_month']=df['bought_in_last_month'].fillna(df['bought_in_last_month'].median())
```

Time for real-work — Feature Engineering
* extracting percentages and amounts and putting them into respective columns to be later used for processing
* columns: `cp%`, `cp$`, `bg%`

```python
df['cp%']=df['is_couponed'].str.extract(r'Save (\d+)%').astype(float)/100
```

```python
df['cp$']=df['is_couponed'].str.extract(r'Save \$(\d+).\d+').astype(float)
```

```python
df['bg%']=df['is_best_seller'].str.extract(r'Save (\d+)%').astype(float)/100
```

last step of fix for `current/discounted_price` 

```python
df['current/discounted_price']=df['current/discounted_price'].str.replace(',','',regex=False)
df['current/discounted_price']=df['current/discounted_price'].astype(float)
```

### creation for bool series (__masks__) for `cp%`, `cp$` and `bg%`
> for `cp%`  
> * where `current/discounted_price` __null__ but `cp%` __not null__  
> * where `of listed_price` __null__  but `cp%` __notnull__  
 
> for `cp$`
> * where `current/discounted_price` __null__ but `cp$` & `listed_price` __not null__
> * where `listed_price` __null__ but `cp$` & `current/discounted_price` __not null__

> for `bg%`
> * where `listed_price` __null__ but `cp$` & `current/discounted_price` __not null__
> * where `current/discounted_price` __null__ but `cp$` & `listed_price` __not null__


```python
current_nulls_n_couponed=(df['current/discounted_price'].isna()) & (df['cp%'].notna())
df[current_nulls_n_couponed]
```

```python
listed_nulls_n_couponed=(df['listed_price'].isna()) & (df['cp%'].notna()) & (df['current/discounted_price'].notna())
df[listed_nulls_n_couponed].head()
```

```python
current_nulls_n_couponed_dollors=(df['cp$'].notna()) & (df['listed_price'].notna()) &(df['current/discounted_price'].isna())
df[current_nulls_n_couponed_dollors]
```

```python
listedp_nulls_n_couponed_dollors=(df['cp$'].notna()) & (df['listed_price'].isna()) & (df['current/discounted_price'].notna())
df[listedp_nulls_n_couponed_dollors].head()
```

```python
listed_nulls_n_bg=(df['bg%'].notna()) & (df['current/discounted_price'].notna()) & (df['listed_price'].isna())
listed_nulls_n_bg.value_counts()
```

```python
current_nulls_n_bg=(df['bg%'].notna()) & (df['current/discounted_price'].isna()) & (df['listed_price'].notna())
current_nulls_n_bg.value_counts()
```

### Application of Masks time  
> _usable masks list_  
> * for `cp%` : `listed_price` __null__ but `cp$` & `current/discounted_price` __notnull__
> * for `cp$` `listed_price` __null__ but `cp$` & `current/discounted_price` __notnull__
> * __mask found no usable rows for__ `bg%`


application of mask where filling `listed_price` by reversing percentage formula to calculate orignal price by divison of `current/discounted_price` / `1-cp%`

```python
df.loc[listed_nulls_n_couponed,'listed_price']=round(df.loc[listed_nulls_n_couponed,'current/discounted_price']/(1-df.loc[listed_nulls_n_couponed,'cp%']),2)
```

Just implaced `current/discounted_price` + `cp$` where mask was true for `listed_price`  
__meaning:__ _I just sum current price with coupon amount and filled where it should belong_

```python
df.loc[listedp_nulls_n_couponed_dollors,'listed_price']=(df.loc[listedp_nulls_n_couponed_dollors,'current/discounted_price']) + (df.loc[listedp_nulls_n_couponed_dollors,'cp$'])
```

cleaning `price_on_variant` with creation of masks where only price amount exists not coupon percentage by using regex pattern `price:\s\$\d+\.\d+$` telling the regex to stop on the last zero of amount  
* for example __price: $12.99__ <- mask stops at exactly 9  
> masks for `price_on_variant` and their logic:
> * First mask is just to cross and verify if $ dosen't exist but % does and where `current/discounted_price` is __null__ __<-__ _empty mask all values are false hehe_😅  
> * where `price_on_variant` contains amount and `current/discounted_price` is __null__ __<-__ _found 8852 truths; time to implace_
> * mask where `listed_price` is __null__ but `current/discounted_price` __not null__ __<-__ _found 16042 rows where `listed_price` is missing yet those values exist in `current/discounted_price`_

```python
df['price_on_variant']=df['price_on_variant'].str.strip()
```

```python
maskre=(~df['price_on_variant'].str.contains(r'price:\s\$\d+\.\d+$')) & (df['price_on_variant'].str.contains(r'price:\s\$\d+\.\d+')) & (df['current/discounted_price'].isna())
df[maskre].head(5)
```

```python
pricever_current=(df['price_on_variant'].str.contains(r'price:\s\$\d+\.\d+$')) & (df['current/discounted_price'].isna())
pricever_current.value_counts()
```
